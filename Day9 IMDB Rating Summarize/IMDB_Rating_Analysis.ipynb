###Task is to find the mean, min, max imdb rating for respective industries namely hollywood and bollywood from given dataset.

```python
import pandas as pd
```

```python
df = pd.read_csv('movies.csv')
df.head(10)
```

```python
df.shape
```

```python
type(df.imdb_rating) #a column in dataframe object is called series.
```

```python
df.imdb_rating.mean(), df.imdb_rating.min(), df.imdb_rating.max()
```

```python
#Filter the dataframe by hollywood, bollywood
bollywood_df =  df[df.industry == 'Bollywood']
hollywood_df =  df[df.industry == 'Hollywood']
```

```python
bollywood_df
```

```python
print("The mean, min, and max ratings for BOLLYWOOD are ",bollywood_df.imdb_rating.mean(), bollywood_df.imdb_rating.min(), bollywood_df.imdb_rating.max())
```

```python
print("The mean, min, and max ratings for HOLLYWOOD are ",hollywood_df.imdb_rating.mean(), hollywood_df.imdb_rating.min(), hollywood_df.imdb_rating.max())
```
